```python

movie_df_path = "processed_data/movies_data.csv"
user_df_path = "processed_data/users_data.csv"
rating_df_path = "processed_data/rating_data.csv"

def preprocess_data(df):
    # Handling null values
    df = df.dropna(how='any')  # Drop rows with any null values
    # Alternatively, you can fill missing values, for example:
    # df['column_name'].fillna(value='default_value', inplace=True)
    # Removing duplicates
    df = df.drop_duplicates()

    return df

# Load the data
movie_df = pd.read_csv(movie_df_path)
user_df = pd.read_csv(user_df_path)
rating_df = pd.read_csv(rating_df_path)

# Preprocess the data
cleaned_movie_df = preprocess_data(movie_df)
cleaned_user_df = preprocess_data(user_df)
cleaned_rating_df = preprocess_data(rating_df)

# Save the cleaned DataFrames back to CSV files
cleaned_movie_df.to_csv(movie_df_path, index=False)
cleaned_user_df.to_csv(user_df_path, index=False)
cleaned_rating_df.to_csv(rating_df_path, index=False)

print("Data preprocessing complete. Cleaned files saved.")
```
